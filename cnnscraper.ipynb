```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
query = 'https://edition.cnn.com/specials/brexit'
out_file= 'outfile.txt'
```

```python
def get_news(query, out_file):
    html = requests.get(query)
    b = soup(html.content, 'lxml')
    #titles
    titles=[]
    for title in b.findAll('h3'):
        titles.append(title.text)
    print(titles)
    #links
    links=[]
    for news in b.findAll('h3', {'class':'cd__headline'},):
        links.append('https://edition.cnn.com' + news.a['href'])
    print(links)
    dates=[]
    news=[]
    for link in links:
        page = requests.get(link)
        bsoup=soup(page.content)
        #snippet of news
        for n in bsoup.findAll('div', {'class':'l-container'}):
            n1=n.text.strip()[:100]
            news.append(n1)
        #date
        for d in bsoup.findAll('p', {'class':'update-time'},):
            dates.append(d.text.strip())
    print(news)
    print(dates)
    #outfile
    with open('out_file.txt', 'w') as filehandle:
        for title in titles:
            filehandle.write('%s\n' % title)
        for link in links:
            filehandle.write('%s\n' % link)
        for n in news:
            filehandle.write('%s\n' % n)
        for date in dates:
            filehandle.write('%s\n' % date)
    with open('out_file.txt', 'r') as filehandle:
        file=filehandle.readlines()
        for line in file:
            current_place=line[:-1]
            titles.append(current_place)
            links.append(current_place)
            news.append(current_place)
            dates.append(current_place)
```

```python
get_news(query, out_file)
```
